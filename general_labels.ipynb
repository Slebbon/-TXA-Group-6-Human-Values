The initial 20 labels will be generalized to only 6, according to the following dictionary

```python
general_dictionary={
    "Self-direction: thought":"Openness to change",
    "Self-direction: action":"Openness to change",
    "Stimulation":"Openness to change",
    "Hedonism":"Openness to change",
    "Achievement":"Self-Enhancement",
    "Power: dominance":"Self-Enhancement",
    "Power: resources":"Self-Enhancement",
    "Face":"Face",
    "Security: personal":"Conservation",
    "Security: societal":"Conservation",
    "Tradition":"Conservation",
    "Conformity: rules":"Conservation",
    "Conformity: interpersonal":"Conservation",
    "Humility":"Humility",
    "Benevolence: caring":"Self-Transcendence",
    "Benevolence: dependability":"Self-Transcendence",
    "Universalism: concern":"Self-Transcendence",
    "Universalism: nature":"Self-Transcendence",
    "Universalism: tolerance":"Self-Transcendence",
    "Universalism: objectivity":"Self-Transcendence"
    
}
```

```python
training_data["Openness to change"]=training_data["Self-direction: thought"]+training_data["Self-direction: action"]+training_data["Stimulation"]+training_data["Hedonism"]
```

```python
training_data["Self-Enhancement"]=training_data["Achievement"]+training_data["Power: dominance"]+training_data["Power: resources"]
```

```python
training_data["Conservation"]=training_data["Security: personal"]+training_data["Security: societal"]+training_data["Tradition"]+training_data["Conformity: rules"]+training_data["Conformity: interpersonal"]
```

```python
training_data["Self-Transcendence"]=training_data["Benevolence: caring"]+training_data["Benevolence: dependability"]+training_data["Universalism: concern"]+training_data["Universalism: nature"]+training_data["Universalism: tolerance"]+training_data["Universalism: objectivity"]
```

Now that the new features are created, the original ones can be dropped

```python
for key in general_dictionary:
    if key!=general_dictionary[key]:
        training_data=training_data.drop(key, axis=1)
    else:
        pass
```

```python
training_data
```

The new labels will now be converted into a probability distribution

```python
# extract the generalized label columns and store them in a separate DataFrame
label_columns=training_data[["Face",
"Humility",
"Openness to change",
"Self-Enhancement",
"Conservation",
"Self-Transcendence"]]

# convert the label columns to a multi-class format (one-hot encoding)
training_data['general label'] = label_columns.apply(lambda row: row.to_list(), axis=1)
```

```python
general_labels=list(training_data["general label"])
```

```python
final=[]

for sublist in general_labels:
    sum_values=sum(sublist)
    probability_distribution=[value/sum_values for value in sublist]
    final.append(probability_distribution)
```

```python
training_data["general label"]=final
```
