```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
path = "arguments-training.tsv"
df = pd.read_table(path, sep = "\t")
df.head()
```

```python
# renaming the Argument ID column so that there is no space
df.rename(columns={"Argument ID": "Argument_ID"}, inplace=True)
```

```python
df.head()
```

```python
l_path = "labels-training.tsv"
temp_labels_df = pd.read_table(l_path, sep="\t")
temp_labels_df.head()
```

```python
# renaming the Argument ID column so that there is no space
temp_labels_df.rename(columns={"Argument ID": "Argument_ID"}, inplace=True)
```

```python
pd.set_option('display.max_columns', None)
temp_labels_df.head()
```

```python
# meking sure that the labels are not unique
labels_dicts = temp_labels_df.to_dict(orient = "records")
for dic in labels_dicts:
    arg = dic["Argument_ID"]
    count = 0
    for key in dic:
        if dic[key] == 1:
            count += 1
    print(f"the argument {arg} has {count} labels ")
```

```python
# tranforming the values of the labels so that if an argument has 2 labels the values are 0.5 each and so on

for dic in labels_dicts:
    count = 0
    values = []
    for key in dic:
        if dic[key] == 1:
            values.append(key)
            count += 1
    for value in values:
        dic[value] = round((1/count), 2)

labels_dicts         
```

```python
# tranforming the labels' dictionary in a dataframe

df_labels = pd.DataFrame.from_dict(labels_dicts)
df_labels
```

```python
# making dure that the labels dataset and the training dataset have the same length
print(f"the number of features in the labels' dataset is: {len(df_labels.columns)}")
print(f"the number of samples in the labels' dataset is: {len(df_labels)}")
print(f"the number of features in the traning dataset is: {len(df.columns)}")
print(f"the number of samples in the training dataset is: {len(df)}")
```

```python
# looking at the distribution od the stances across the training dataset
l_favourable = len(df.loc[df["Stance"]  == "in favor of"])
l_against = len(df.loc[df["Stance"]  == "against"])
favourable_perc = round(l_favourable/(l_favourable + l_against) * 100, 2)
against_perc = round(l_against/(l_favourable + l_against) * 100, 2)
print(f"The arguments with an 'in favour' stance are {l_favourable}, the {favourable_perc}%, meanwhile the arguments 'against' are {l_against}, the {against_perc}%")

lab = ["In favour", "Against"]
l_array = np.array([l_favourable, l_against])
plt.pie(l_array, labels=lab)
plt.title("Distribution of the Stance")
plt.legend()
plt.show()
```
